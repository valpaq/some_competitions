Категориальных значений немного, можно построить boxplot и посмотреть на распределения.

```python
fig,ax=plt.subplots(len(obj),figsize=(20,70))
for i,ob in enumerate(obj):
    sns.boxplot(data=train,x=ob,y='y',ax=ax[i])
```

как можно увидеть, есть определенные зависимости от категориальной переменных и распределений предсказываемой переменной:

```python
train.describe()
```

Как можно увидеть из обычного describe, есть переменная x11, в которой все 0. А сколько таких еще? И вдруг есть какие-нибудь не только с 0, а со всеми 1?

```python
mas=train.dtypes[train.dtypes=='int64'].index[1:]
nan = []
for i in mas:
    if (train[i].var()==0):
        print(i, train[i].var(),train[i][0])
        nan.append(i)
```

Ага, надо их удалить

```python
train = train.drop(columns=nan, axis=1)
```

```python
train.describe()
```

У очень многих переменных std и mean меньше 1, плюс судя по describe у многих медиана равна 0, интересно сколько их и какие они

```python
mas=train.dtypes[train.dtypes=='int64'].index[1:]
strange = []
for i in mas:
    if (train[i].std()<=1) and (train[i].mean()<=1):
        print(i, train[i].std(),train[i].mean())
        strange.append(i)
```

Таких очень много, как будто было всего 5 переменных и сделали one-hot
